```python
import numpy as np
import pandas as pd
import pandas_profiling
import seaborn as sns
```

```python
invoices = pd.read_csv('data/invoices.csv')
customers = pd.read_csv('data/customers.csv')

invoices['created_at'] = pd.to_datetime(invoices["created_at"])
invoices['due'] = pd.to_datetime(invoices["due"])
invoices['last_payment_date']= pd.to_datetime(invoices["last_payment_date"])
invoices = invoices.drop(['currency', 'exchange_rate', 'paid', 'source_system'], axis=1)

customers['created_at'] = pd.to_datetime(customers["created_at"])
customers['created_at'] = customers['created_at'].apply(lambda x: x - pd.DateOffset(years=1))
customers = customers.set_index('customer_id')
```

```python
pandas_profiling.ProfileReport(invoices)
```

```python
pandas_profiling.ProfileReport(customers)
```

```python
df = pd.merge(invoices, customers, left_on='customer_id', right_index=True, suffixes=('', '_customer'))
```
